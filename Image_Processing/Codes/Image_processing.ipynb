## Image Rotating

```python
imageT = cv2.imread('Resorces_Images/park.jpg',0)
```

```python
w,h=imageT.shape
```

```python
center=(w // 2, h // 2)
```

```python
M = cv2.getRotationMatrix2D(center, -90, 1.0)
rotated = cv2.warpAffine(imageT, M, (w, h))
cv2.imshow("Rotation", rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Filtering(Gaussian-Median-Bilateral)  

```python
image= cv2.imread('Resorces_Images/nemo3.jpg')
```

```python
# Gaussian Blur
gauss = cv2.GaussianBlur(image, (11,11), 0)
cv2.imshow('Gaussian Blur', gauss)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Averaging
average = cv2.blur(image, (11,11))
cv2.imshow('Average Blur', average)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Median Blur
median = cv2.medianBlur(image, 3)
cv2.imshow('Median Blur', median)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Bilateral
bilateral = cv2.bilateralFilter(image, 10, 35, 25)
cv2.imshow('Bilateral', bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Edge detection(Canny)  

```python
image= cv2.imread('Resorces_Images/nemo3.jpg')
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges=cv2.Canny(gray,100,200,apertureSize=3)
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(image)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(edges)
```
